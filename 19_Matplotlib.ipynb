```
import numpy as np
import matplotlib.pyplot as plt
```

```
%matplotlib inline 
```

```
age = [10,20,30,30,30,40,50,60,70,75]
weight = [20,60,80,85,86,87,70,90,95,90]
numpyAge = np.array(age)
numpyWeight = np.array(weight)
```

```
plt.plot(numpyAge,numpyWeight, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs. Weight")
plt.show()
```

```
numpyArray = np.linspace(0,10,20)
plt.plot(numpyArray,"g*-")
```

```
numpyArray2 = numpyArray **3
plt.plot(numpyArray2,"r*-")
```

```
plt.subplot(1,2,1)
plt.plot(numpyArray,numpyArray2,"r*-")
plt.subplot(1,2,2) # plotting second graph
plt.plot(numpyArray2,numpyArray,"g--")
```

```
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.4,0.4,0.9,0.9])
figureAxes.plot(numpyArray, numpyArray2,"g")
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Title")
```

```
numpyArray1 = np.linspace(0,10,20)
numpyArray2 = numpyArray1 ** 2
(myFigure, myAxis) = plt.subplots()
myAxis.plot(numpyArray,numpyArray2,"g--",linewidth = 2.0)
myAxis.plot(numpyArray2,numpyArray,"b:",linewidth = 2.0)
```
